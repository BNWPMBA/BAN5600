```python
# Question 1
count = 0
total_sum = 0
product = 1
sum_squares = 0

num = 150
while num <= 1000:
    if num % 2 != 0:
        count += 1
        total_sum += num
        product *= num
        sum_squares += num**2
    num += 1

print(f"a) Count of odd numbers: {count}")
print(f"b) Sum of odd numbers: {total_sum}")
print(f"c) Multiplication of odd numbers: {product}")
print(f"d) Sum of squares of odd numbers: {sum_squares}")
```

```python
# Question 2
count = 0
total_sum = 0
product = 1
sum_squares = 0

for num in range(150, 1001): # 1001 to include 1000
    if num % 2 != 0:
        count += 1
        total_sum += num
        product *= num
        sum_squares += num**2

print(f"a) Count: {count}")
print(f"b) Sum: {total_sum}")
print(f"c) Product: {product}")
print(f"d) Sum of squares: {sum_squares}")
```

```python
# Question 3
M = int(input("Enter value for M: "))
s = 0
k = 1

while k <= M:
    s = s + (1/k)
    k = k + 1

print(f"The sum s is: {s}")
```

```python
# Question 4
total = 0
count = 0

while True:
    val = input("Enter a number: ")
    if val.lower() == "done":
        break
    try:
        num = float(val)
        total += num
        count += 1
    except ValueError:
        print("bad data Invalid input")

if count > 0:
    print(f"{total} {count} {total/count}")
else:
    print("No numbers were entered.")
```

```python
# Question 5
nums = []

while True:
    val = input("Enter a number: ")
    if val.lower() == "done":
        break
    try:
        num = float(val)
        nums.append(num)
    except ValueError:
        print("bad data Invalid input")

if nums:
    print(f"Total: {sum(nums)}, Count: {len(nums)}, Max: {max(nums)}, Min: {min(nums)}")
else:
    print("No numbers were entered.")
```
