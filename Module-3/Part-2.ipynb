```python
import time
from google import genai
from google.colab import userdata

# 1. Setup the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

# 2. Define the most current stable model for 2026
# If 2.5-flash fails, 'gemini-2.0-flash' is the fallback
model_id = "gemini-2.5-flash"

def ask_gemini(prompt):
    try:
        response = client.models.generate_content(
            model=model_id,
            contents=prompt
        )
        print(f"--- Response from {model_id} ---")
        print(response.text)
    except Exception as e:
        if "429" in str(e):
            print("Quota hit! Waiting 20 seconds before retrying...")
            time.sleep(20) # Wait for the 'per-minute' limit to reset
            return ask_gemini(prompt) # Try one more time
        else:
            print(f"An error occurred: {e}")

# 3. Run the request
ask_gemini("Explain how to use a list in Python.")
```
