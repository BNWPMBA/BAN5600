```python
# --- PROBLEM 5: Merging Reports from Multiple Departments ---
# Objective - Write a Python workflow that combines two related datasets into one unified report.

customers = [
    {"ID": 1, "Name": "Alpha Co", "Region": "North"},
    {"ID": 2, "Name": "Beta Inc", "Region": "South"},
    {"ID": 3, "Name": "Gamma LLC", "Region": "East"},
    {"ID": 4, "Name": "Delta Group", "Region": "West"}
]

orders = [
    {"Order_ID": 5001, "Cust_ID": 1, "Amount": 1200},
    {"Order_ID": 5002, "Cust_ID": 2, "Amount": 450},
    {"Order_ID": 5003, "Cust_ID": 1, "Amount": 700},
    {"Order_ID": 5005, "Cust_ID": 99, "Amount": 300} # Ghost order
]

merged_report = []
total_processed = 0

# Use a while loop to iterate through customers for the 'Merge'
while total_processed < len(customers):
    cust = customers[total_processed]
    found_order = False

    # Use a for loop to look for matching orders (The "Inner Join" logic)
    for order in orders:
        if order["Cust_ID"] == cust["ID"]:
            entry = {
                "Name": cust["Name"],
                "Amount": order["Amount"],
                "Status": "Active"
            }
            merged_report.append(entry)
            found_order = True
            # We found a match, continue searching for more orders for this customer
            continue

    if not found_order:
        print(f"Alert: Customer {cust['Name']} has no matching orders.")

    total_processed += 1

# Calculate total revenue manually
grand_total = 0
print("\n--- FINAL SALES REPORT ---")
for record in merged_report:
    print(f"Customer: {record['Name']} | Sale: ${record['Amount']}")
    grand_total += record["Amount"]

    # Break logic: Stop summary if we exceed a budget threshold for this report
    if grand_total > 5000:
        print("Budget threshold reached. Stopping summary.")
        break

print(f"Total Combined Revenue: ${grand_total}")
```

```python
# --- Problem 7. Organizing Customer or Client Email Responses
# Objective - Create a Python notebook that generates personalized email response drafts based on request type and customer details.


client_requests = [
    {"Name": "Alice", "Type": "Appointment", "Status": "Active", "Date": "2024-05-01"},
    {"Name": "Bob", "Type": "Payment", "Status": "Past Due", "Date": "2024-04-10"},
    {"Name": "Charlie", "Type": "Appointment", "Status": "Active", "Date": "2024-05-02"},
    {"Name": "Dana", "Type": "Shipping", "Status": "Active", "Date": "2024-04-20"},
    {"Name": "Unknown", "Type": "Inquiry", "Status": "Inactive", "Date": "N/A"}
]

outbox = []
i = 0

# Use a while loop to process the email queue
while i < len(client_requests):
    req = client_requests[i]
    i += 1

    # Skip processing for "Unknown" or "Inactive" accounts (continue)
    if req["Status"] == "Inactive":
        print(f"Skipping record for {req['Name']}: Account Inactive.")
        continue

    # Logic to build the message (if/elif/else)
    if req["Type"] == "Appointment":
        msg = f"Hi {req['Name']}, your appointment on {req['Date']} is confirmed!"
    elif req["Type"] == "Payment":
        urgency = "URGENT" if req["Status"] == "Past Due" else "Friendly"
        msg = f"Hello {req['Name']}, this is a {urgency} reminder for your payment."
    elif req["Type"] == "Shipping":
        msg = f"Dear {req['Name']}, your order is on the way as of {req['Date']}."
    else:
        msg = f"Hello {req['Name']}, thank you for your inquiry."

    outbox.append(msg)

    # Stop processing if the outbox hits a limit (break)
    if len(outbox) >= 4:
        print("Outbox full. Stopping batch process.")
        break

# Display the generated emails using a for loop
print("\n--- GENERATED EMAIL DRAFTS ---")
for email in outbox:
    print(f"Draft: {email}")

# Basic calculation for metrics
success_rate = (len(outbox) / len(client_requests)) * 100
print(f"\nAutomation Success Rate: {success_rate}%")
```

```python
# --- Problem 8. Monitoring Shipment Delays and Missing Order Information
# Objective - Use Python to review shipment records, flag problem orders, and produce an exception report.


shipments = [
    {"ID": "ORD01", "Status": "Delivered", "Days": 5, "Docs": "Yes"},
    {"ID": "ORD02", "Status": "In Transit", "Days": 2, "Docs": "Yes"},
    {"ID": "ORD03", "Status": "Delayed", "Days": 10, "Docs": "No"},
    {"ID": "ORD04", "Status": "Delivered", "Days": 4, "Docs": "Yes"},
    {"ID": "ORD05", "Status": "Lost", "Days": 15, "Docs": "No"}
]

audit_log = []
index = 0

# Use a while loop to process the shipment queue
while index < len(shipments):
    current = shipments[index]
    index += 1

    # Skip shipments that are still in transit (continue)
    if current["Status"] == "In Transit":
        print(f"Skipping {current['ID']}: Still in transit.")
        continue

    # Stop the audit if we hit a 'Lost' shipment for investigation (break)
    if current["Status"] == "Lost":
        print(f"CRITICAL ERROR: {current['ID']} is LOST. Stopping audit.")
        break

    # Logic to determine if a shipment is an 'Exception'
    is_exception = False
    if current["Days"] > 7:
        is_exception = True
    elif current["Docs"] == "No":
        is_exception = True

    # If it's an exception, add it to our report list
    if is_exception:
        report_entry = [current["ID"], current["Status"], "FLAGGED"]
        audit_log.append(report_entry)
    else:
        print(f"Shipment {current['ID']} passed audit.")

# Print summary using a for loop
print("\n--- FINAL EXCEPTION REPORT ---")
for item in audit_log:
    print(f"Order: {item[0]} | Status: {item[1]} | Action: {item[2]}")

# Calculation to meet basic math requirement
total_shipments = len(shipments)
exception_rate = (len(audit_log) / total_shipments) * 100
print(f"\nException Rate: {exception_rate}%")
```

```python
# --- Problem 10. Tracking Task Deadlines and Follow-Ups
# Objective - Develop a Python notebook that reads task data, identifies upcoming and overdue work, and generates a reminder report.


tasks = [
    ["Project Alpha", "Alice", 2],  # Name, Assigned, Days Overdue
    ["Compliance Audit", "Bob", -5], # Negative means days remaining
    ["Client Call", "Charlie", 0],
    ["Budget Review", "Alice", 10],
    ["Monthly Report", "David", 1]
]

high_priority_reminders = []

print("--- ANALYZING TEAM WORKLOAD ---")
for task in tasks:
    task_name = task[0]
    owner = task[1]
    days_past = task[2]

    # Check if task is overdue using if statements
    if days_past > 0:
        status = "OVERDUE"
        # Nested logic: If very overdue, add to priority list
        if days_past > 7:
            high_priority_reminders.append(task_name)
    elif days_past == 0:
        status = "DUE TODAY"
    else:
        status = "ON TRACK"

    print(f"Task: {task_name} | Assigned to: {owner} | Status: {status}")

# Process the priority list with a while loop and control flow
print("\n--- PROCESSING URGENT REMINDERS ---")
count = 0
while count < len(high_priority_reminders):
    reminder = high_priority_reminders[count]

    if reminder == "Budget Review":
        print(f"SENDING URGENT EMAIL to Alice for {reminder}")

    count += 1

# Final list manipulation (Array requirement)
team_members = ["Alice", "Bob", "Charlie", "David", "Eve"]
if "Eve" not in [t[1] for t in tasks]:
    print("\nNote: Eve has no tasks assigned currently.")
```
